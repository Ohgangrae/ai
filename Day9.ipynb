```python
# (영어)알파벳과 숫자를 섞어서 입력을 받아서 숫자와 문자를 
# 분리하여 출력하는 프로그램을 작성하시오.

# [입력]
# 숫자를 포함한 문자열을 입력하세요 >>> 123abc45

# [출력]
# 문자 : abc
# 숫자 : 12345

str = input('숫자를 포함한 문자열을 입력하세요 >>> ')

al = ''
no = ''

for i in str:
  if i.isalpha():
    al = al + i
  elif i.isdigit():
    no = no + i

print('문자 : ', al)
print('숫자 : ', no)
```

```python
# 문자와 숫자의 개수를 각각 포함하여 출력하세요

# 숫자를 포함한 문자열을 입력하세요 >>> 1234asdf324as
# 문자(6) :  asdfas
# 숫자(7) :  1234324


str = input('숫자를 포함한 문자열을 입력하세요 >>> ')

al = ''
no = ''

for i in str:
  if i.isalpha():
    al = al + i
  elif i.isdigit():
    no = no + i

print(f'문자({len(al)}) : {al}')
print(f'숫자({len(no)}) : {no}')
```

```python
# by 김현엽

n=input("입력값>>")
a=""
b=""
c=0
d=0
for i in n:
    if i in "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ":
        c=c+1
        b=b+i    
    elif i in '0123456789':
        d=d+1
        a=a+i
    
        
print(f"문자{c}: {b}")     
print(f"숫자{d}: {a}")
```

```python
# by 김정관
a = input("숫자를 포함한 문자열을 입력하세요.>>>")
n = ''
al = ''
al_list = []
n_list = '1234567890'

# 알파벳 리스트
for i in range(97,123) :
  al_list.append(chr(i))

for i in range(65,91) :
  al_list.append(chr(i))




for i in a :
  if i in al_list :
    al = al + i
  elif i in n_list :
    n = n + i

print(f'문자({len(al)}) : {al}')
print(f'숫자({len(n)}) : {n}')
```

```python
# by 이지은

str = input('숫자를 포함한 문자열을 입력하세요 >>> ')
cnt1 = 0
al = ''
cnt2 = 0
no = ''

for i in str:
  if i == '1' or i == '2' or i == '3' or i == '4' or i == '5' or i == '6' or i == '7' or i == '8' or i == '9' or i == '0':
    cnt2+=2
    no+=i
  else:
    cnt1+=1
    al+=i

print(f'문자({cnt1}):{al}')
print(f'숫자({cnt2}):{no}')
```

```python
# by 이재원
def str_slide( pr ):
    s = "1234567890"
    a = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    str = "";
    int = "";
    count_int = 0; 
    count_str = 0;
    
    for i in pr:
        if i in s:
            int += f"{i} ";
            count_int +=1 ;
        elif i in a:
            str += f"{i} ";
            count_str += 1;            
            
    print(f"입력 숫자 : {int} {count_int}개");
    print(f"입력 문자 : {str} {count_str}개");
            
pr = input("입력 > ");
str_slide(pr);
```

```python
# by 임진석

str1 = input('알파벳과 숫자를 입력 : ')
al = ''
num = ''
cnt_al = 0
cnt_num = 0
for i in str1:
    if i >='A' and i <= 'Z':
        al += i
        cnt_al += 1
    elif i >='a' and i <= 'z':
        al += i
        cnt_al += 1
    elif i >= '0' and i <='9':
        num += i
        cnt_num += 1

        
print(f'문자({cnt_al}) : {al}')
print(f'숫자({cnt_num}) : {num}')
```

```python
# by 민지수

str=input('숫자를 포함한 문자열을 입력하세요>>> ')

ai=""
no=""

for i in str :
    if 64<ord(i)<123:
        ai=ai+i
    
    elif 48<ord(i)<58:
        no=no+i
    
    
print(f'문자({len(ai)}) : ',ai)
print(f'숫자({len(no)}): ',no)
```

```python
########################################################################
# 알고리즘
########################################################################
# '알고리즘은 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한
# 아이디어나 생각'
# 알고리즘 표현 방법은 수도코드(pseudo code), 순서도(flow chart) 이 있다. 

# 프로그램 작성 순서
# 기획 - 설계(알고리즘) - 프로그래밍 - 디버그 - 문서작성


# 좋은 알고리즘
# 알기 쉽고, 속도가 빠르고, 효율적이고(메모리를 적게사용), 재이용이 쉽다. 

# 왜 알고리즘 ??
# 좋은 프로그램 만들수 있고, 좋은 프로그램인지 판단할수 있고 
# 프로그램 작성을 효율적으로 할수 있고 프로그래밍 기술을 향상시키는데 큰 도움이 된다. 

# 좋은 알고리즘의 조건
# 결과를 얻을수 있어야 하고 반드시 종료되어야 한다. 

# 알고리즘 세가지 기본형
#1 순차구조 : 처음부터 순서대로 처리
#2 선택구조 : 조건식 if
#3 반복구조 : 반복문 for while



# 순서도 flow 차트, 흐름도
# 프로그래밍 언어를 사용하지 않고 알고리즘을 기술하는 방법
# 처리와 정차를 시각적으로 표현 한다. 
# 자신이 생각하고 있는 알고리즘이 올바른지, 처리와 절차에 문제가 없는지를
# 눈으로 확인할수 있고 
# 주의 사항
# 처리 기호나 판단 기호가 많아지면 흐름이 복잡하게 꼬인다. 
# 1. 입력은 반드시 위에서 들어가도록 만든다. 
# 2. 판단기호로 부터 나오는 흐름선에는 Yes no를 확실하게 기입한다. 
# 3. 입력선이나 출력선이 없는 처리 기호는 만들지 않는다. 


```
